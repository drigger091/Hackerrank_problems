```python
def longestPalindrome(a): 
    
  s = len(a)
  res = ""
  reslen = 0

  if s <=0:
      print("Cant be done")

  else:

      for i in range(s):

        if s %2 != 0: #odd length
            l,r = i,i
            while l >=0 and r < s and a[l] == a[r]:
              if (r-l +1) > reslen:
                res = a[l:r+1]
                reslen = r-l +1
              l-=1
              r+=1
            else: #even length
              l ,r = i ,i+1

            while l >=0 and r < s and a[i] == a[r]:

              if (r-1+l) > reslen:
                res = a[l:r+1]
                reslen = l-r +1
              l-=1
              r+=1

      return res


print(longestPalindrome(a))
```
